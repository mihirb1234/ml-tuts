IMPORTING DEPENDENDIES

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

label encoding of breast cancer dataset


```python
#loading data 
cancer_data=pd.read_csv('/content/cancer_datset.csv')
cancer_data.head()
```

```python
#finding the count of different data
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encoder=LabelEncoder()
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

appending the labels to the dataframe

```python
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
# malignant means cancer present M so we encoded it to 1 as malignant
```

```python
cancer_data['target'].value_counts()
```
